```python
def get_birds(lat, long, region_name,bird_code='baleag' ):
    birds = requests.get(f'https://api.ebird.org/v2/data/nearest/geo/recent/{bird_code}/?lat={lat}&lng={long}', headers=headers)
    return  birds.json()
```

```python
crow_central_park = get_birds('40.7812', '-73.9665', 'Central Park', 'amecro')
crow_loop = get_birds('41.8786', '-87.6251', 'Loop', 'amecro')
crow_seattle = get_birds('47.6062', '-122.3321', 'Seattle', 'amecro')
```

```python
# Create a histogram of the number of crows per sighting
```

```python
crow_counts = []

crow_locs = [crow_central_park, crow_loop, crow_seattle]

counts = []
for loc in crow_locs:
    print(loc)
    counts = [sighting['howMany'] for sighting in loc]
    print(counts)
```

```python
def get_how_many(location):
    counts = []
    for sighting in location:
        try:
            counts.append(sighting['howMany'])
        except:
            continue
        
    return counts
```

```python
cp_crows = get_how_many(crow_central_park)
loop_crows = get_how_many(crow_loop)
seattle_crows = get_how_many(crow_seattle)
```

In the lists above, each element represents a number of crows seen in a given sighting at a given location.

First, combine the lists, and create a histogram showing the distribution of number of crows per sighting across all locations.

```python
#__SOLUTION__
import numpy as np
all_crows = cp_crows + loop_crows + seattle_crows

fig, ax = plt.subplots()

ax.hist(all_crows, bins=100 );
```

The histogram you plotted probably looks pretty unintersting.  That is because there are some outlier bird sightings with a large number of birds sighted.
